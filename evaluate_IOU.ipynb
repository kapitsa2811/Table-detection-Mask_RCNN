```python
# import the necessary packages
from collections import namedtuple
import numpy as np
import cv2
import pandas as pd
# define the `Detection` object
Detection = namedtuple("Detection", ["image_path", "gt", "pred"])
examples = list()
def bb_intersection_over_union(boxA, boxB):
    # determine the (x, y)-coordinates of the intersection rectangle
    xA = max(boxA[0], boxB[0])
    yA = max(boxA[1], boxB[1])
    xB = min(boxA[2], boxB[2])
    yB = min(boxA[3], boxB[3])

    # compute the area of intersection rectangle
    interArea = max(0, xB - xA + 1) * max(0, yB - yA + 1)

    # compute the area of both the prediction and ground-truth
    # rectangles
    boxAArea = (boxA[2] - boxA[0] + 1) * (boxA[3] - boxA[1] + 1)
    boxBArea = (boxB[2] - boxB[0] + 1) * (boxB[3] - boxB[1] + 1)

    # compute the intersection over union by taking the intersection
    # area and dividing it by the sum of prediction + ground-truth
    # areas - the interesection area
    iou = interArea / float(boxAArea + boxBArea - interArea)

    # return the intersection over union value
    return iou
# pred_box =list()
# image_name =list()
# countt =0
# bbox_image_count =0
# annotations = pd.read_csv(os.path.join(ROOT_DIR,"samples/table/"+"test.csv"))
# gt_values= list()
        
# # define the list of example detections
# #for img in pred_bbox:
# for index,row in annotations.iterrows(): #iterating train.csv by rows
#     gt_values.clear()
#     gt_values.append(row[2])
#     gt_values.append(row[1])
#     gt_values.append(row[4])
#     gt_values.append(row[3])
#     pred_box.clear()
#     if row[0] in image_name:
#         bbox_image_count+= 4
#     else:
#         bbox_image_count = 0
#     for i in range(bbox_image_count,bbox_image_count+4):
#         if i < len(pred_bbox[row[0]]):
#             pred_box.append(pred_bbox[row[0]][i])
#     #            print(pred_bbox[row[0]][i])
#     countt+=1

        
#     examples.append(Detection(row[0], [point for point in gt_values], [point for point in pred_box]))
#     image_name.append(row[0])
```
